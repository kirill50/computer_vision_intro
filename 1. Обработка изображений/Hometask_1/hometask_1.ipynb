```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import imageio
from skimage.color import rgb2hsv, hsv2rgb
```

```python
img = imageio.imread('cat-color.png')
img = img.astype(np.float32) / 255.
print(img.shape)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
for i in range(img.shape[2]):
    if i == 1: #RGB - red green blue
        img[:,:,i] = img[:,:,i]**1.5
    if i == 0:
        img[:,:,i] = img[:,:,i]**0.5
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
for c in range(3):
    img[0:700,650:750,c] = 0
    img[0:700,0:100,c] = 0
    img[0:50,0:750,c] = 0
    img[600:700,0:750,c] = 0
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img)
```

```python
#transform image fro RGB to HSV
img_hsv = rgb2hsv(img)
#tune colors intensity
img_hsv[:, :, 2] /= 1.5
img_new = hsv2rgb(img_hsv)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(img_new)
```
